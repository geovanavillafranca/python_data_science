```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
# Load boston dataset
boston_real_state_data = load_boston()
```

```python
print(boston_real_state_data)
```

```python
# Define x axis for the data
x_axis = boston_real_state_data.data
```

```python
# Define y axis for the target
y_axis = boston_real_state_data.target
```

```python
# Plot Histogram
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.hist(y_axis, bins=50)
plt.xlabel('price in 1000s US')
plt.ylabel('number of houses')
plt.show()
```

```python
# Plot scatter plot
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(boston_real_state_data.data[:,5], boston_real_state_data.target)
plt.ylabel('price in 1000s')
plt.xlabel('number')
```
